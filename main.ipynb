```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data_set/customer_churn.csv')
df.head()
```

```python
df.info()
```

# Replacing blanks with 0 and converting datatype to float 

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', '0')
df['TotalCharges'] = df['TotalCharges'].astype('float')
```

```python
df.info()
```

# Checking null values

```python
df.isnull().sum() # returns the sum of null values in each column
df.isnull().sum().sum() # returns overall sum of null values in the dataset
```

```python
df.describe()
```

# Checking for duplicate values based on cutomer id

```python
df['customerID'].duplicated().sum()
```

# Converting 0 and 1 values of senior citizen to yes / no

```python
def conv(value): 
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

```python
ax = sns.countplot(x = df['Churn'], data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (6,6))
gb = df.groupby('Churn').agg({'Churn' : 'count'})
plt.pie(gb['Churn'], labels = gb.index , autopct = "%.1f%%" )
plt.title("Percentage of Churned Customers")
plt.show()
```

```python
sns.countplot(x = df['gender'], data = df , hue = 'Churn')
plt.title('Churn by Gender')
plt.show()
```

```python
sns.countplot(x = df['SeniorCitizen'], data = df , hue = 'Churn')
plt.title('Churn by Senior Citizen')
plt.show()
```

```python
 churn_counts = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

churn_percentage = churn_counts.div(churn_counts.sum(axis=1), axis=0) * 100
ax = churn_percentage.plot(kind='bar', stacked=True, figsize=(8, 6))

for p in ax.patches:
    height = p.get_height()
    width = p.get_width()
    x = p.get_x() + width / 2
    y = p.get_y() + height / 2
    ax.annotate(f'{height:.1f}%', (x, y), ha='center', va='center')

plt.title('Churn by Senior Citizen (Stacked Bar with Percentages)', fontsize=16)
plt.xlabel('Senior Citizen (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (9, 4))
sns.histplot(x = 'tenure', data = df, bins = 72 ,hue = 'Churn')
plt.show()
```

```python
ax = sns.countplot(x = 'Contract', data = df, hue = 'Churn')
ax.bar_label(ax.containers[0])
plt.title("Churn by Contracts")
plt.show()
```

```python
columns =  ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
           'StreamingTV', 'StreamingMovies']

fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))  

axes = axes.flatten()

for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
